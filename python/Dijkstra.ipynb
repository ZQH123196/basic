
迪杰斯特拉算法，其实就是在 BFS 的基础上进行的改进。

因为原本 BFS 算法是假定所有节点间的距离是相等的，任意节点间的距离相等，那就相当于不用考虑距离，所以 BFS 是在不考虑距离的情况下的最短路径算法。

如果要考虑路径或者说要考虑权重之类的东西，各个节点间的通行代价不一样了，就不能使用 BFS 了，要用迪杰斯特拉算法。

这个算法也简单，对 BFS 的队列进行改进将起点跟当前节点的距离加进去考虑就行了，同时对除来源外可通行的全部节点每次都重新判定。

1. 去除每个节点只访问一次的保证



只要改两点：
1. 将队列改为优先队列。
2. 始终考虑起点到当前节点的距离。

> 优先队列是指将参考值作为基准，自动从小到大排序参考值的队列，相比普通队列多了自动按照参考值排序的功能。

优先队列有个参考值，因为没有参考值如何体现优先呢？自然是要有个值进行大小的比较，才能体现出优先。
此处，我们都会以小的值为优先，因为是求最短路径嘛。

以前的队列，我们只要放入节点即可，但是改成了优先队列，我们还需要一个能够体现优先级的值。
在迪杰斯特拉算法中，是以距离（权重）作为参考值进行计算的，优先队列会用这个值来进行排序。

队列：队列保存一个值，这个值由一个节点的位置构成。
优先队列：优先队列保存一个值，这个值由一个节点+这个节点到起点的距离构成。
> 就相当于在结构体中除了坐标外在多加了一个值，可以说相当简单了。

那么，这个当前节点到起点的距离是怎么算的呢？

首先，我们是从起点出发，然后将可通行的节点与“自身到这些节点的距离跟自身的距离相累加后的值”拼接起来，然后存到到优先队列，然后优先队列根据距离值排序一番。
第二，我们从优先队列弹出第二个节点，然后将可通行的节点与“自身到这些节点的距离跟自身的距离相累加后的值”拼接起来，然后优先队列根据距离值排序一番。
...

注意这里的“自身到这些节点的距离跟自身的距离相累加后的值”，这句话的意思是每一次存放距离时，不仅要算上当前节点到下一节点的距离，还要跟当前节点自身保存的距离值进行相加。
而当前节点保存的距离值是什么呢？其实是起点到当前节点的距离。

距离值是不断累加的，由于是从起点开始，因此每个节点当前的距离都是起点到自身的距离，所以当前个节点的下一个可通行节点的距离 == 起点到当前节点的距离 + 当前节点到下一可通行节点的距离。
